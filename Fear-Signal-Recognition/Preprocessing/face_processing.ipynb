```python
#Libraries used

import subprocess
import os
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report, accuracy_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
#Neural Network
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Path to the single video file, test file replace with the file you want to test
video_path = '../Test_Video/Test_video.mp4'

# Directory where you want to save the output files
output_folder = '../Test_Video'

# Ensure the output directory exists
os.makedirs(output_folder, exist_ok=True)

# Construct the feature extraction command, this uses OpenFace
cmd = f'./FeatureExtraction -f "{video_path}" -out_dir "{output_folder}"'

# Execute the command
os.system(cmd)

print("Feature extraction completed for the video.")

# Path to the directory containing your CSV files
directory_path = '../Test_Video'

# Placeholder for the selected data
selected_data = []

# Iterate over each file in the directory
for filename in os.listdir(directory_path):
    if filename.endswith('.csv'):
        # Construct the full file path
        file_path = os.path.join(directory_path, filename)
        
        # Read the CSV file
        df = pd.read_csv(file_path)
        
        df['fear_score'] = df['AU01_r'] + df['AU02_r'] + df['AU04_r'] - df['AU06_r'] + df['AU20_r'] + df['AU25_r']
        # Select the row with the maximum 'fear_score'
        selected_row = df.loc[df['fear_score'].idxmax()]
        
        # Add the filename (without extension) and selected row data to the list
        selected_data.append([os.path.splitext(filename)[0]] + selected_row.tolist())

# Convert the list to a DataFrame
selected_df = pd.DataFrame(selected_data, columns=['File Name'] + list(df.columns))

# Save the selected DataFrame as a new CSV file
output_csv_path = '../Test_Video/face_csv_file.csv'
selected_df.to_csv(output_csv_path, index=False)

print(f'Selected data CSV saved to {output_csv_path}')
```

### Transforming the combined .csv

```python
# Load your data
data = pd.read_csv("./processed/FinalData/OpenFace_data.csv")

# Function to map file name prefixes to countries
def map_country(file_name):
    if file_name.startswith('CA'):
        return 'Canada'
    elif file_name.startswith('TR'):
        return 'Turkey'
    elif file_name.startswith('C'):
        return 'China'
    elif file_name.startswith('I'):
        return 'India'
    else:
        return 'Unknown'

# Function to extract emotion from file name
def extract_emotion(file_name):
    parts = file_name.split('_')  # Split the file name by '_'
    if len(parts) > 1:
        emotion = parts[1].lower()  # Get the emotion part and convert to lower case
        if emotion in ['nervous', 'threatened', 'surprised']:
            return emotion
    return 'unknown'  # Return 'unknown' if no valid emotion is found

# Apply functions to create new columns
data['Country'] = data['File Name'].apply(map_country)
data['Emotion'] = data['File Name'].apply(extract_emotion)

# Save the updated DataFrame
output_path = "./Classification_csv/Face_features.csv"
data.to_csv(output_path, index=False)

print(f"Updated data with 'Country' and 'Emotion' columns saved to {output_path}")
```
